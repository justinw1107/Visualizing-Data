```r
# install.packages("tidyverse")
# install.packages("treemap")
# install.packages("dplyr")

library(tidyverse)
library(dplyr)
library(treemap)
```

```r
datapath = file.path("data", "Olympics.csv")
OlympicsCSV = read.csv(datapath)
head(OlympicsCSV)
```

```r
# Dropping unneeded columns
OlympicsCSV = subset(OlympicsCSV, select = -c(Name, Height, Weight, Team, Games, City))
head(OlympicsCSV)
```

```r
# Checking for NA values
sum(is.na(OlympicsCSV$ID))
sum(is.na(OlympicsCSV$Sex))
sum(is.na(OlympicsCSV$Age))
sum(is.na(OlympicsCSV$NOC))
sum(is.na(OlympicsCSV$Year))
sum(is.na(OlympicsCSV$Season))
sum(is.na(OlympicsCSV$Sport))
sum(is.na(OlympicsCSV$Event))
sum(is.na(OlympicsCSV$Medal))
```

```r
# Dropping all rows with NA values in any column
OlympicsCSV <- na.omit(OlympicsCSV)
sum(is.na(OlympicsCSV))
```

```r
# Counting medals per NOC
MedalsByNoc <- OlympicsCSV %>% 
  group_by(NOC, Medal) %>% 
  summarise(count = n()) %>%
  mutate(Medal = factor(Medal, levels = c("Gold", "Silver", "Bronze")))

# Aggregate medal count
MedalsByNocAgg <- aggregate(count ~ NOC, data = MedalsByNoc, FUN = sum)

# Sort by descending order
MedalsByNocAgg <- MedalsByNocAgg[order(MedalsByNocAgg$count, decreasing = TRUE),]

# Create column with desired label
MedalsByNocAgg$label <- paste(MedalsByNocAgg$NOC, MedalsByNocAgg$count, sep = "\n");

# Create treemap
treemap(MedalsByNocAgg, 
        index = "label", 
        vSize = "count",
        type = "index",
        title = "Total Medals won in Olympic History per NOC",
        fontsize.labels = 10,
        title.legend = "Medal"
)
```

```r
# Counting medals per NOC
MedalsByNoc <- OlympicsCSV %>%
  group_by(NOC, Medal) %>%
  summarise(count = n()) %>%
  mutate(Medal = factor(Medal, levels = c("Gold", "Silver", "Bronze"))) %>%
  group_by(NOC) %>%
  summarise(TotalMedals = sum(count))

# Get top 5 NOCs
Top5NOCs <- MedalsByNoc %>%
  arrange(desc(TotalMedals)) %>%
  head(5)

# Create new dataframe with top 5 NOCs
OlympicsCSVTop5 <- OlympicsCSV %>%
  filter(NOC %in% Top5NOCs$NOC)

# Sort new dataframe by medal count
MedalsByNocTop5 <- OlympicsCSVTop5 %>%
  group_by(NOC, Medal) %>%
  summarise(count = n()) %>%
  mutate(Medal = factor(Medal, levels = c("Gold", "Silver", "Bronze")))

# Color vector
medal_colors <- c("Gold" = "#FFD700", "Silver" = "#C0C0C0", "Bronze" = "#CD7F32")

# Create stacked bar chart
ggplot(MedalsByNocTop5, aes(x = NOC, y = count, fill = Medal)) +
  geom_bar(stat = "identity") +
  scale_fill_manual(values = medal_colors) + 
  labs(title = "Medal Distribution for the Top 5 NOCs",
       x = "National Olympic Committee",
       y = "Number of Medals",
       fill = "Medal Type") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1),
        legend.position = "bottom",
        legend.title = element_blank())
```

```r
# Get most dominant NOC by counting medals
MostDominatingNOC <- MedalsByNoc %>%
  filter(TotalMedals == max(TotalMedals)) %>%
  pull(NOC)

# Create new dataframe with most dominant NOC
OlympicsCSVMostDominating <- OlympicsCSV %>%
  filter(NOC == MostDominatingNOC)

# Sort new dataframe by medals per sport
MedalsBySportMostDominating <- OlympicsCSVMostDominating %>%
  group_by(Sport, Medal) %>%
  summarise(count = n()) %>%
  mutate(Medal = factor(Medal, levels = c("Gold", "Silver", "Bronze")))

# Create new dataframe with top 10 best sports
MedalsBySportTop10 <- MedalsBySportMostDominating %>%
  group_by(Sport) %>%
  summarise(TotalMedals = sum(count)) %>%
  top_n(10, TotalMedals) %>%
  inner_join(MedalsBySportMostDominating, by = "Sport")

# Color vector
medal_colors <- c("Gold" = "#FFD700", "Silver" = "#C0C0C0", "Bronze" = "#CD7F32")

# Create the horizontal stacked bar chart
ggplot(MedalsBySportTop10, aes(y = reorder(Sport, count), x = count, fill = Medal)) +
  geom_bar(stat = "identity") +
  scale_fill_manual(values = medal_colors) +
  labs(title = paste0(MostDominatingNOC, "'s most winning Olympic Sports"),
       x = "Number of Medals",
       y = "Sport",
       fill = "Medal Type") +
  theme_minimal() +
  theme(axis.text.y = element_text(hjust = 1),
        legend.position = "bottom",
        legend.title = element_blank())
```

```r
# Get most dominant NOC by counting medals
MostDominatingNOC <- MedalsByNoc %>%
  filter(TotalMedals == max(TotalMedals)) %>%
  pull(NOC)

# Create new dataframe with most dominant NOC
OlympicsCSVMostDominating <- OlympicsCSV %>%
  filter(NOC == MostDominatingNOC)

# Count the number of male and female athletes
AthleteCounts <- OlympicsCSVMostDominating %>%
  group_by(Sex) %>%
  summarise(count = n())

# Calculate percentages of each sex
AthleteCounts$percent <- scales::percent(AthleteCounts$count / sum(AthleteCounts$count))

# Create the pie chart
ggplot(AthleteCounts, aes(x = "", y = count, fill = Sex)) +
  geom_bar(stat = "identity", color = "white") +
  coord_polar("y", start=0) +
  theme_void() +
  geom_text(aes(label = percent), position = position_stack(vjust = 0.5), size = 8) +
  scale_fill_manual(values = c("#FFC0CB", "#87CEEB"), labels = c("Female", "Male")) +
  labs(fill = "Sex") +
  labs(title = paste0("Gender Distribution of ", MostDominatingNOC, "'s Medal Winners")) +
  theme(legend.text = element_text(size = 12), legend.title = element_text(size = 14))
```
